Utilizando o Python para obter o spreed entre ações ordinárias e preferenciais.

Importando algumas libs e capturar os dados dos pares PETR4/PETR3, depois
BBDC4/BBDC3 e CMIG4/CMIG3

```python
#Instalando e importando Libs

!pip install investpy
import investpy as inv
import plotly.express as px
```

```python
#Obtendo dados dos ativos:
# PETR4 e PETR3

petr4 = inv.get_stock_historical_data('PETR4', country='brazil', from_date='01/01/2022', to_date='12/09/2022')

petr3 = inv.get_stock_historical_data('PETR3', country='brazil', from_date='01/01/2022', to_date='12/09/2022')
```

Calcular essa métrica é importante pois nos dá uma noção de como os papéis estão esticados com relação ao seu par.

Para o calculo de uma razão, simplesmente dividimos o preço de uma pela outra e alocamos o resultado numa nova coluna (imagem)

```python
#calculo da razão PETR:

long_short = petr4.Close/petr3.Close
long_short.head()
#long_short
```

Quando a razão se estica muito _(se distancia da média)_, **maiores são as chances de voltar à média.**

Vamos criar um plot utilizando a biblioteca _Plotly_ para entender como essa razão se comportou ao longo do tempo

```python
# Criação do Gŕafico:

fig = px.line(x=long_short.index, y=long_short)

fig.add_hline(y=long_short.mean(), line_width=5, line_color="green")

fig.add_hline(y=(long_short.mean()-long_short.std()), line_width=3, line_dash="dash",line_color="orange")

fig.add_hline(y=(long_short.mean()+long_short.std()), line_width=3, line_dash="dash", line_color="orange")

fig.add_hline(y=(long_short.mean()-2*long_short.std()), line_width=5, line_dash="dash", line_color="red")

fig.add_hline(y=(long_short.mean()+2*long_short.std()), line_width=5, line_dash="dash", line_color="red")

fig.update_layout(xaxis_rangeslider_visible=False, title_text='Razão entre preço PETR4 e PETR3(01/2022 e set/2022)', template = 'simple_white', width = 500, height=500)

fig.show()
```
